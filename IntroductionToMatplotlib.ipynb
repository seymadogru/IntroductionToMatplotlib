Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3 π  and 3 π  with 0.1 intervals.

```python
import numpy as np
import matplotlib.pyplot as plt
π=3.14
x=np.arange(-3*π,3*π  , 0.1)
y=np.cos(x)
plt.plot(x,y,color="g")
plt.show()
```

This time, create the same plot in red color with object methods.

```python
import numpy as np
import matplotlib.pyplot as plt
myFigure , myAxes = plt.subplots()
π=3.14
x=np.arange(-3*π,3*π  , 0.1)
y=np.cos(x)
myAxes.plot(x,y,color="#B24069")
plt.show()
```
